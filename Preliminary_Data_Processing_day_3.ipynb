being the midprice between `Buy` and `Sell`


```python
def getMidPrices(df):
    buy_prices = df['bid_price_1']
    sell_prices = df['ask_price_1']
    mid_prices = []
    for i in range(0,len(df)):
        mid_prices.append((int(buy_prices[i]) + int(sell_prices[i]))/2)
    return mid_prices

dg_midprices_shifted = getMidPrices(diving_gear_copy) # should be midpricesETF
dg_midprices= getMidPrices(diving_gear) 
ds_midprices = dolphin_sightings['mid_price']

fig, ax1 = plt.subplots(figsize=(14,14))
color = 'tab:red'
ax1.set_xlabel('timestamp')
ax1.set_ylabel('diving gear', color=color)
ax1.plot(dg_midprices_shifted, color=color)
ax1.tick_params(axis='y', labelcolor=color)

ax2 = ax1.twinx()  # instantiate a second axes that shares the same x-axis

color = 'tab:blue'
ax2.set_ylabel('dolphin sightings', color=color)  # we already handled the x-label with ax1
ax2.plot(ds_midprices, color=color)
ax2.tick_params(axis='y', labelcolor=color)

'''fig.tight_layout()  # otherwise the right y-label is slightly clipped
plt.show()
plt.plot(np.array(dg_midprices),label= 'Diving Gear 3 days')
plt.plot(np.array(ds_midprices),label= 'Dolphin Sightings 3 days')
plt.legend()'''
```

```python
import numpy as np
from scipy import signal
y = dg_midprices
x = ds_processed
correlation = signal.correlate(x-np.mean(x), y - np.mean(y), mode="full")
lags = signal.correlation_lags(len(x), len(y), mode="full")
lag = lags[np.argmax(correlation)]
```

```python
def berryMidPrices(df):
    buy_prices = df['bid_price_1'].values.tolist()
    buy_prices = [eval(i) for i in buy_prices]
    sell_prices = df['bid_price_1'].values.tolist()
    sell_prices = [eval(i) for i in sell_prices]
    print(buy_prices)
    print(sell_prices)
    mid_prices = (np.array(buy_prices )+ np.array(sell_prices))/2

    return mid_prices
```
